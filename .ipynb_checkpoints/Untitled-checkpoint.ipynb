# Section One - Data Loading

```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Reading data

df = pd.read_csv('housing.csv')
```

```python
#Getting first 5 rows

print(df.head())
```

# Section Two - Data Clean and Processing

```python
#Gettings info about datas

print(df.info())
```

```python
#Handling missing values

median_total_bedrooms = df['total_bedrooms'].median()
df['total_bedrooms'].fillna(median_total_bedrooms, inplace=True)
```

```python
#Coding of categorical variables

df_encoded = pd.get_dummies(df, columns=['ocean_proximity'], prefix=['ocean'])
```

```python
#Splitting the data into training and testing sets

X = df_encoded.drop('median_house_value', axis=1)
y = df_encoded['median_house_value']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Importing library

from sklearn.preprocessing import StandardScaler
```

```python
#Data normalization


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Section Four - Prediction and Creating Machine Learning Models

```python
#Importing libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error
```

```python
#Splitting the data into a training and a test set

encoder = LabelEncoder()
X['ocean_proximity_encoded'] = encoder.fit_transform(X['ocean_proximity'])

X_encoded = X.drop('ocean_proximity', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
#Creating model for LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
linear_pred = linear_model.predict(X_test)
linear_mse = mean_squared_error(y_test, linear_pred)
```

```python
#Creating model for DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
tree_pred = tree_model.predict(X_test)
tree_mse = mean_squared_error(y_test, tree_pred)
```

```python
#Printing Mean Squared Error

print('Linear Regression Mean Squared Error:', linear_mse)
print('Decision Tree Mean Squared Error:', tree_mse)
```

```python
#Visualization for the LinearRegression

sns.scatterplot(x=y_test, y=linear_pred)
sns.lineplot(x=y_test, y=y_test, color='red', label='Ideal Line')
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
plt.title('Linear Regression - Predicted vs True Values')
plt.show()
```

```python
#Visualization for the DecisionTreeClassifier

sns.scatterplot(x=y_test, y=tree_pred)
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
plt.title('Decision Tree - Predicted vs True Values')
plt.show()
```
